```
import numpy as np
from keras.preprocessing import image
from keras.models import Model, Sequential
from keras.layers import Dense, Dropout
from keras.applications import VGG19
```

```
model = VGG19(include_top = True,weights = 'imagenet')
model.summary()
```

```
def preprocessing(img_path):
  img = image.load_img(img_path, target_size = (224,224,3))
  im = image.img_to_array(img)
  im = np.expand_dims(im , axis = 0)
  pred = model.predict(im)
  return pred
  
```

```
path = '/content/drive/My Drive/dataset'
```

```
import os
```

```
dict = {}
x_data = []
y_data = []
count = 0
for folder in os.listdir(path):
  dict[count] = folder
  for file in os.listdir(path + '/' + folder):
    if(file[0] != '.' and file[len(file) - 2] != 'v'):
      x_data.append(preprocessing(path + '/' + folder + '/' + file))
      y_data.append(count)
  print(count)
  count = count + 1
```

```
X = np.array(x_data)
Y = np.array(y_data)
print(X.shape, Y.shape)
num_classes = np.unique(Y).shape[0]
from keras.utils import np_utils
Y = np_utils.to_categorical(Y)
```

```
x = np.squeeze(X, axis = 1)
print(Y.shape)
```

```
model_new = Sequential()
model_new.add(Dense(512, input_dim = (1000), activation='relu'))
model_new.add(Dropout(0.3))
model_new.add(Dense(256, activation='relu'))
model_new.add(Dropout(0.3))
model_new.add(Dense(num_classes, activation='softmax'))
print(model_new.summary())
model_new.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```
hist = model_new.fit(x,Y,shuffle=True, epochs=40, batch_size=2,validation_split = 0.2)
```

```
model_new.save('/content/drive/My Drive/model.h5')
model_new.save_weights('/content/drive/My Drive/model_weights.h5')
```
